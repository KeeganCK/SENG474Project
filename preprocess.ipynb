```python
import pandas as pd
import sklearn
df=pd.read_csv("nodes.csv")
df=df.sort_values('popularity',ascending=False)
df=df[0:1500]
df.head()
# df.to_csv('out.csv')
```

```python
import itertools
df['genres'] = df['genres'].str.strip('[]').str.replace("'","").str.split('\s*,\s*')
chartFreq={}
for index, row in df.iterrows():
    chart=df.at[index,'genres']
    for item in chart:
        if item in chartFreq:
      # incerementing the frequency count by 1
            chartFreq[item]+=1
        else:
            # setting the count to 1
            chartFreq[item]=1
        
        
# df1 = (
#     df['genres'].explode()
#     .str.get_dummies().sum(level=0).add_prefix('genre_')
# )
# df1 = df.drop('genres', 1).join(df1)
sorted_chartFreq = sorted(chartFreq.items(), key=lambda x:x[1], reverse=True)
converted_dict = dict(sorted_chartFreq)
df=df.drop(columns=['chart_hits'])
converted_dict=dict(itertools.islice(converted_dict.items(), 25))
print(converted_dict)
```

```python
import numpy as np
genre=[]
for key in converted_dict:
    genre.append(key)
df["new_genres"] = np.empty((len(df), 0)).tolist()
print(df.dtypes)
for index, row in df.iterrows():
    chart=df.at[index,'genres']
    new_genre=[]
    for item in chart:
        if item in genre:
            new_genre.append(str(item))
    if not new_genre:
        df.drop(index,axis=0, inplace=True)
    else:
        print(new_genre)
        df.at[index,"new_genre"]=new_genre
        # df.at[index,'new_genre']+=new_genre
print(df.head())
```
